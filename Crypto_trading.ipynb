```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
trades = pd.read_excel("historical_data.xlsx")
fear = pd.read_excel("fear_greed_index.xlsx")
```

```python
trades.head()
fear.head()
```

```python
trades.columns = trades.columns.str.strip().str.lower().str.replace(" ", "_")
fear.columns = fear.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
trades.head()
```

```python
trades.info()
```

```python
#trades['timestamp"] = pd.to_datetime(trades[timestamp], unit = 'ms')
trades['timestamp'] = pd.to_datetime(trades['timestamp'], unit='ms')
fear["date"] = pd.to_datetime (fear['timestamp'], unit = "s")
```

```python
trades.head()
```

```python
trades = trades.dropna()
fear = fear.dropna()
```

```python
trades['date'] = trades['timestamp'].dt.date
fear['date'] = fear['date'].dt.date
```

```python
trades['trade_value'] = trades['execution_price'] * trades['size_tokens']
```

```python
trades['pnl_status'] = np.where(trades['closed_pnl'] > 0, 'profit', 'loss')
```

```python
merged = pd.merge(trades, fear, on='date', how='left')
```

```python
#merged.head()
```

```python
merged = merged.drop(columns = ['timestamp_y'])
```

```python
#merged.head()
```
